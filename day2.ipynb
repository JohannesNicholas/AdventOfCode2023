```python
# get the data from the clipboard
data = pd.read_clipboard(header=None, sep="zxdgasdfasdf")
display(data)
```

```python
#part 2

# Add a column for game ID
data["game_id"] = data[0].str.extract(r"(\d+)")

# remove Game x:


total = 0

# Loop through the rows
for index, row in data.iterrows():

    min_possible = {
        "red" : 0,
        "green": 0,
        "blue": 0,
    }
    
    clean = row[0].replace("Game " + row["game_id"] + ": ", "")
    print(clean)

    for round_idx, round in enumerate(clean.split(";")):
        
        for color in round.split(","):
            #print(color)
            number, color = color.strip().split(" ")

            #print(f"Round {round_idx} - {color} - {number}")
            
            if min_possible[color] < int(number):
                min_possible[color] = int(number)

    print(min_possible)

    power = int(min_possible["red"]) * int(min_possible["green"]) * int(min_possible["blue"])

    print(power)

    total+=power

print(total)
```
