```python
import pandas as pd
```

```python
# get the data from the clipboard as a string
data = pd.read_clipboard(header=None)
display(data)
```

```python
total = 0

# loop through each line and find the first and last number digit
for line in data[0]:
    # split the line into a list of strings
    line = line.split()[0]
    print(line)

    

    for char in line:
        # check if the string is a number
        if char.isdigit():
            print(char)
            first = char
            break

    for char in line[::-1]:
        # check if the string is a number
        if char.isdigit():
            print(char)
            last = char
            break
    
    number = first + last

    total += int(number)

print(total)


```

```python
total = 0

numbersAsWords = [
    'zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine'
]

# loop through each line and find the first and last number digit
for line in data[0]:
    # split the line into a list of strings
    line = line.split()[0]
    print(line)

    

    for i, char in enumerate(line):
        # check if the string is a number
        if char.isdigit():
            print(char)
            first = char
            break

        #check if the next characters make a number word
        found = False
        for word in numbersAsWords:
            if line[i:i+len(word)] == word:
                print(word)
                first = str(numbersAsWords.index(word))
                found = True
                break
        
        if found:
            break

    for ib, char in enumerate(line[::-1]):
        i = len(line) - ib - 1
        # check if the string is a number
        if char.isdigit():
            print(char)
            last = char
            break


        found = False
        #check if the next characters make a number word
        for word in numbersAsWords:
            if line[i:i+len(word)] == word:
                print(word)
                last = str(numbersAsWords.index(word))
                found = True
                break
        
        if found:
            break
    
    number = first + last

    print(number)

    total += int(number)

print(total)
# copy just the total to the clipboard
pd.DataFrame([total]).to_clipboard(index=False, header=False)


```
