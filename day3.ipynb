```python
# get the data from the clipboard
input = pd.read_clipboard(header=None, sep="sdfhsfhgsfhg")
display(input)
```

```python


# Loop through the grid pixel by pixel

# convert the dataframe to a 2d python list
inp = [row[0] for row in input.values.tolist()]

inp 

total = 0

numberAsString = '0123456789'


for row_i, row in enumerate(inp):

    current_number = ''

    

    for char_i, char in enumerate(row):

        if char in '0123456789':
            current_number += char
        

        if char not in '0123456789' or char_i == len(row) - 1:
            if current_number != '':
                

                noSymbol = True

                #search the surroundings for symbols
                for row_search_i in range(row_i - 1, row_i + 2):
                    for char_search_i in range(char_i - len(current_number) - 1, char_i + 1):

                        if row_search_i < 0 or row_search_i >= len(inp):
                            continue
                        if char_search_i < 0 or char_search_i >= len(inp[row_search_i]):
                            continue
                        if inp[row_search_i][char_search_i] in current_number:
                            continue

                        if inp[row_search_i][char_search_i] == '.':
                            continue

                        
                        if not noSymbol:
                            print("AAAHHHHHHGGGGG")

                        print('\033[92m' + current_number + '\033[0m', end='')

                        total += int(current_number)

                        noSymbol = False
                        

                if noSymbol:
                    #print the number in red
                    print('\033[91m' + str(int(current_number)) + '\033[0m', end='')


                current_number = ''

        if char not in '0123456789':
            print(char, end='')


    
    print(f"    {total}")


print()
print()
print(total)

```
